```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('arch.jpeg.jpeg')
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(imgRGB, cv2.COLOR_RGB2GRAY)
img_blur = cv2.GaussianBlur(img_gray, (3, 3), 0)
plt.title("Original Image")
plt.imshow(imgRGB)
```

```python
#HARRIS CORNER DETECTION
corner_img = cv2.cornerHarris(img_gray, 2, 5, 0.07)
corner_img = cv2.dilate(corner_img, None)
img[corner_img > 0.01 * corner_img.max()]=[0, 0, 255]
plt.imshow(img)
```
